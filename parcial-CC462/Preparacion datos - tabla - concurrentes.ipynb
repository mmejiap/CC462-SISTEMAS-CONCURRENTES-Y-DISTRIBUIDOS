```python
# etiquetar grupo de cada 30 min desde 04:45 a 23:45
fecha="2018-05-04"
esp=" "
hora_inicio="04:45:00"
hora_fin="23:45:00"

#w_bloque=dt.timedelta(minutes=30)
w_bloque=30

#fh_inicio=dt.datetime.strptime(fecha+esp+hora_inicio,'%Y-%m-%d %H:%M:%S')
#fh_fin=dt.datetime.strptime(fecha+esp+hora_fin,'%Y-%m-%d %H:%M:%S')

fh_inicio=dt.datetime.strptime(hora_inicio,'%H:%M:%S')
fh_fin=dt.datetime.strptime(hora_fin,'%H:%M:%S')

etiqueta_inicial=0

column_etiqueta_hora=[]

vector = data01['HORAS_format']

num_grupos=0

temp_i=fh_inicio
bloques_tiempo=[]
bloques_tiempo.append(temp_i.time())
while(temp_i<fh_fin):
    #print("->",temp_i)
    num_grupos=num_grupos+1
    temp_i=fh_inicio+dt.timedelta(minutes=w_bloque*(num_grupos))
    bloques_tiempo.append(temp_i.time())
    #print("->",temp_i)
    #print("---", num_grupos)


jj=0
print(bloques_tiempo)
for i in range(len(vector)):
    k=0
    while k < (len(bloques_tiempo)-1):
        ti=bloques_tiempo[k]
        tf=bloques_tiempo[k+1]
        if(vector[i]>ti and vector[i]<=tf):
            column_etiqueta_hora.append(k)
        #else:
        #    jj=jj+1
        k=k+1
    
print("grupos:",column_etiqueta_hora)
```
